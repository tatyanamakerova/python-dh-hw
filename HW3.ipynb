## Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
#устанавливаем библиотеку 
from pymystem3 import Mystem
#определяем функцию для лемматизации и задаем аргументы
def lemmatization(first_text, second_text):
    m = Mystem()
    #задаем условия,ключевые аргументы и кодировку 
    open(file=second_text, mode='w', encoding='UTF-8') 
    #задаем условия, ключи и кодировку для исходного файла
    with open(file=first_text, mode='r', encoding='UTF-8') as original:
        for line in original:
            lemmas = m.lemmatize(line) 
            #задаем условия, ключи и кодировку для лемматизированного файла
            with open(file=second_text, mode='a', encoding='UTF-8') as lemmatizated:
                lemmatizated.write(''.join(lemmas))
#определяем функцию для исходного и лемматизированного файлов
def main():
    first_text_path = '/Users/tatyanamakerova/Desktop/prog/literary_anecdotes.txt'
    second_text_path = '/Users/tatyanamakerova/Desktop/prog/lemmatized_literary_anecdotes.txt'
    lemmatization(first_text_path, second_text_path) #вызываем функцию, которая лемматизирует файл    
if __name__ == '__main__':
    main() #вызываем функцию 
```

## Задание №3

Сделайте полный морфологический разбор исходного текста. Напишите регулярное выражение, которое будет извлекать из тега только часть речи. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). Посчитайте абсолютную частоту для всех частей речи, а затем относительнную частоту (абсолютная частота / длина текста в словах).

```python
#открываем текст и читаем его 
with open(r'/Users/tatyanamakerova/Desktop/prog/lemmatized_literary_anecdotes.txt', 'r', encoding='utf-8') as f:
    il_testo = f.read()
#импортируем модуль и делаем морфологический анализ текста
from pymystem3 import Mystem 
m = Mystem()
morfologia = m.analyze(il_testo) 
#импортируем модуль с регулярными выражениями
import re
#компилируем регулярку для поиска частей речи
rex = re.compile('[A-Z]+') 
#создаем пустой словарь
dizionario = {} 
#проходим циклом по словарю и списку с морфологическим анализом 
for il_nuovo_diz in morfologia: 
    for key, value in il_nuovo_diz.items():
        try: #вводим специальную конструкцию для обработки исключений
            forma = il_nuovo_diz['text']  #text-слово, полученное на выход 
            parte_il_discorso = il_nuovo_diz['analysis'][0]['gr'] #в словаре на выходе должен быть анализ 'analysis' и грамматическая информация 'gr' 
            #помним, что форма -- ключ, часть речи -- значение - используем регулярное выражение
            dizionario[forma] = rex.match(parte_il_discorso).group(0) 
        except (KeyError, IndexError) as e: #пишем названия ошибок
            pass #задаем условие, которое будет их учитывать
#распечатываем словарь
print(dizionario) 
#достаем из словаря все значения и записываем их в список
senso = dizionario.values()
#считаем частотность
from collections import Counter
counts = Counter(senso)
print (counts)
#для относительной частотности используем lambda
new_counts = {}
for key, value in counts.items():
    ipm = value / len(counts.values()) * 1000000
new_counts[key] = ipm 
print(new_counts) #распечатываем 
  
```
